```python
t_70, n_events = [], 10000
mean_events = 61 * 200                  # approximate integral rate of events >70 ADC in SSD
random_mip = np.random.normal(loc=10, size=n_events)

for mip in random_mip:

    sample_size = np.random.poisson(mean_events)         
    peaks = np.random.exponential(mip, size=sample_size)
    peaks = np.sort(peaks)

    t_70.append(peaks[500])             # idk random index
```
