```python
import matplotlib.pyplot as plt
import numpy as np
import sys, os
```

```python
station = "NuriaJr"
data_dir = f"/cr/tempdata01/filip/SSDCalib/UubRatesT1/{station}"
hits = np.zeros(350)

for file in os.listdir(data_dir):
    hits += np.loadtxt(f"{data_dir}/{file}")

def get_best_threshold(hits, target):
    #      t = files n_tr    t_bin   n_bin
    duration = 800 * 5000 * 8.33e-9 * 2048
    rates = hits / duration

    return np.argmin(np.abs(rates * 100/rates[0]- target))
```

```python
print(get_best_threshold(hits, 100))
print(get_best_threshold(hits, 70))
print(get_best_threshold(hits, 20))
```

```python
duration = 800 * 5000 * 8.33e-9 * 2048

rates = hits/duration * 100/(hits[0]/duration)

plt.plot(range(350), rates)
```
