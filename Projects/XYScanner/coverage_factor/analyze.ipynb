```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
x, y = [], []

for f in os.listdir():
    if not f.endswith('txt'): continue

    cal_xy_corr = np.loadtxt(f, usecols=4)

    x.append(float(f[:-4]))
    y.append(np.mean(cal_xy_corr))

    
    plt.boxplot(cal_xy_corr, positions=[x[-1]], widths=[0.03])


plt.xlim(0.25, 1.15)

popt = np.polyfit(x, y, 1)
plt.plot(x, np.poly1d(popt)(x))

plt.xlabel('coverage factor')
plt.ylabel('XY calib. constant')
```
