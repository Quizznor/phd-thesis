```python
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.colorbar import ColorbarBase
from mpl_toolkits.axes_grid1.inset_locator import mark_inset
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
import numpy as np

fig, ((ax1, ax2), (cax1, cax2)) = plt.subplots(2, 2, figsize=(6.6, 3.3), height_ratios=[1.0, 0.05])

dx, dy = 0.015, 0.05
y, x = np.mgrid[slice(-4, 4 + dy, dy), 
                slice(-4, 4 + dx, dx)] 
z = (1 - x / 3. + x ** 5 + y ** 5) * np.exp(-x ** 2 - y ** 2) 
z = z[:-1, :-1] 
z_min, z_max = -np.abs(z).max(), np.abs(z).max() 
  
ax1.imshow(z, cmap ='Greens', vmin = z_min, vmax = z_max, 
                 extent =[x.min(), x.max(), y.min(), y.max()], 
                    interpolation ='nearest', origin ='lower') 

ColorbarBase(cax1, cmap='Greens', 
             norm=Normalize(vmin=0, vmax=1), 
             orientation='horizontal')
ax2.set_xlim(-20, 20)
ax2.set_ylim(-20, 20)

ax_zoomed = zoomed_inset_axes(ax2, 4, loc=1)
mark_inset(ax2, ax_zoomed, loc1=2, loc2=4, fc="none", ec="k", lw=0.4)
ax_zoomed.set_xticklabels([])
ax_zoomed.set_yticklabels([])

ax_zoomed.scatter(10, 10)

# ax_zoomed.axhline((OFFSET_Y + MAX_DISPLACEMENT_Y)*1.e-3, ls='-.', c='k', lw=0.2)
# ax_zoomed.axhline((OFFSET_Y - MAX_DISPLACEMENT_Y)*1e-3, ls='-.', c='k', lw=0.2)
# ax_zoomed.axvline((MAX_DISPLACEMENT_X)*1e-3, ls='-.', c='k', lw=0.2)
# ax_zoomed.axvline((-MAX_DISPLACEMENT_X)*1e-3, ls='-.', c='k', lw=0.2)
# ax_zoomed.fill_between([-lim, -MAX_DISPLACEMENT_X*1e-3], -lim, lim, color='r', alpha=0.5, edgecolor="None")
# ax_zoomed.fill_between([MAX_DISPLACEMENT_X*1e-3, lim], -lim, lim, color='r', alpha=0.5, edgecolor="None")
# ax_zoomed.fill_between([-MAX_DISPLACEMENT_X*1e-3, MAX_DISPLACEMENT_X*1e-3], -lim, 
#                        (OFFSET_Y - MAX_DISPLACEMENT_Y)*1e-3, color='r', alpha=0.5, edgecolor='None')
# ax_zoomed.fill_between([-MAX_DISPLACEMENT_X*1e-3, MAX_DISPLACEMENT_X*1e-3], (OFFSET_Y + MAX_DISPLACEMENT_Y)*1e-3, 
#                        lim, color='r', alpha=0.5, edgecolor='None')
# ax_zoomed.fill_between([-MAX_DISPLACEMENT_X*1e-3, MAX_DISPLACEMENT_X*1e-3], (OFFSET_Y - MAX_DISPLACEMENT_Y)*1e-3,
#                        (OFFSET_Y + MAX_DISPLACEMENT_Y)*1e-3, color='g', alpha=0.5, edgecolor='None')
# ax_zoomed.scatter(mean_x / 1000, mean_y / 1000, marker='*', s=10, c='k', label='average all')
                        
cax2.axis('off')
```
