```python
from utils.plotting import *
from datetime import datetime

fig, axes = plt.subplots(7, 1, sharex=True, sharey=True)

for id, ax in zip(names.keys(), axes):

    DAratio = DayTest('fDynodeAnode', int(id)).T
    time = DayTest('fTime', int(id))

    ax.scatter([datetime.fromtimestamp(t) for t in time], DAratio[0])
    ax.scatter([datetime.fromtimestamp(t) for t in time], DAratio[1])
    ax.scatter([datetime.fromtimestamp(t) for t in time], DAratio[2])

    ax.set_ylabel(f'Station: {names[id]} ({id})', rotation=0, labelpad=30, fontsize=4)

plt.ylim(28, 37)
# plt.scatter([],[], label='PMT 1')
# plt.scatter([],[], label='PMT 2')
# plt.scatter([],[], label='PMT 3')
# fig.legend(loc='outside upper right')
```
