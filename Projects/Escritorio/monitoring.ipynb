```python
from Monit import Monit

from datetime import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import dates

def format(ax):
    locator = dates.AutoDateLocator()
    formatter = dates.ConciseDateFormatter(locator)
    ax.xaxis.set_major_formatter(formatter)
```

```python
monit = Monit(2025, 1, range(7,9))
```

```python
stations = [609, 943, 944, 949, 954]
```

```python
for station in stations:
    time = monit('fTime', station)[1:]
    peak = monit('fPeak[4]', station)[1:]

    time = [dt.fromtimestamp(t) for t in time]

    plt.plot(time, peak[:, -1], label=station)

format(plt.gca())
plt.legend()
```
