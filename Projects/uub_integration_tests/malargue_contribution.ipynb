```python
from utils.binaries import *
from utils.plotting import *
from utils.auger.sd.moni import Monit

def do_shit(monit: Monit) -> plt.Figure:
    
    # monit and plotting results
    fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)
    # fig, axes = plt.subplots(1, 2, sharex=True)
    # colors = iter(plt.rcParams['axes.prop_cycle'].by_key()['color'])
    colors = iter(['k', 'r', 'lime', 'c', 'g', 'b', 'slategray'])
    linestyle = iter(['--', ':', '-', "-.", "--", "-", ":"])
    for station in [943, 944, 945, 948, 949, 954, 972]:

        if station in [943, 944, 949, 954]:
            ax = ax1
        else:
            ax = ax2
        
        c = next(colors)
        ls = next(linestyle)

        time = monit('fTime', station)
        mip = monit('fPeak', station)[:, -1]
        # rate = monit('f70HzRate', station)

        ax.plot(plot.to_datetime(time[1:]), mip[1:], marker='none', label=station, c=c, ls=ls)

    ax1.legend(ncol=4)
    ax2.legend(ncol=3)
    plot.apply_datetime_format(ax2)
    ax1.set_ylabel(r"$I^\mathrm{SSD}_\mathrm{rate}$ / ADC", labelpad=12)
    ax2.set_ylabel(r"$I^\mathrm{SSD}_\mathrm{rate}$ / ADC")

    ax1.set_ylim(20, 35)
    ax2.set_ylim(0, 10)
        
```

```python
do_shit(Monit(2025, 2, range(21,24)))
```
