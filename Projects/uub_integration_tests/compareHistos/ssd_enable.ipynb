```python
from utils.binaries import *
from utils.plotting import *
```

```python
v_before = "v22"
v_ssd_enable = "v23"


def get_histo(version: str) -> np.ndarray:
    path = f"/cr/data01/filip/Data/daqIntegrationTests/{version}/out.txt"
    offset = 4 if version == "v22" else 0

    data = np.loadtxt(path, skiprows=offset)

    return data[:4, 4:]
```

```python
histo_before = get_histo(v_before)
histo_ssd_enable = get_histo(v_ssd_enable)
```

```python
from utils import CONSTANTS

fig, axes = plt.subplots(nrows=2)

for i, ax in enumerate(axes, 2):
    bins = CONSTANTS.UUB_WCD_BIN_CENTERS if i != 3 else CONSTANTS.UUB_SSD_BIN_CENTERS
    ax.plot(bins, histo_before[i], label="before")
    ax.plot(bins, histo_ssd_enable[i], label="ssd_enable")

    ax.legend()
```
